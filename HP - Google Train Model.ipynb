```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from itertools import product
import warnings

# Load the dataset
df = pd.read_csv('GOOG.csv') # Replace 'your_dataset.csv' with the path to your dataset file

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Set 'date' column as index
df.set_index('date', inplace=True)

# Select only 'close' column for time series analysis
ts = df['close']

# Train-test split
train = ts[:'2020']
test = ts['2021']

# Define the range of p, d, and q values
p_values = range(0, 5) # Adjust the range as needed
d_values = range(0, 5) # Adjust the range as needed
q_values = range(0, 5) # Adjust the range as needed

# Generate all different combinations of p, d, and q triplets
pdq = list(product(p_values, d_values, q_values))
best_aic = float("inf")
best_pdq = None

# Grid search over all possible combinations of p, d, and q
for param in pdq:
    try:
        with warnings.catch_warnings():
            warnings.filterwarnings("ignore")
            # Fit ARIMA model
            model = ARIMA(train, order=param)
            results = model.fit()
            
        # Calculate AIC
        aic = results.aic
        
        # Check if current model is better than the best found so far
        if aic < best_aic:
            best_aic = aic
            best_pdq = param
    except:
        continue
        
# Fit the best model
best_model = ARIMA(train, order=best_pdq)
best_results = best_model.fit()

# Forecast
forecast = best_results.forecast(steps=len(test))

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(test.index, test.values, label='Actual')
plt.plot(test.index, forecast, label='Predicted')
plt.title('Google Close Price Prediction for 2021')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()

# Print best model parameters and AIC
print(f"Best model parameters: ARIMA{best_pdq}")
print(f"Best AIC: {best_aic}")
```
